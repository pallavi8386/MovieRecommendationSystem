```python
# we need only name feature from genres. Use the 'ast' package to fetch it.
```

```python
import ast
```

```python
def fetchVal(text):
    L = []
    for i in ast.literal_eval(text):
        L.append(i['name'])
    return L
```

```python
movies['genres'] = movies['genres'].apply(fetchVal)
```

```python
movies.head()
```

```python
#Now we see just the names list in Genres column. Likewise, we need to extract just names from keywords feature
```

```python
movies['keywords']=movies['keywords'].apply(fetchVal)
```

```python
movies.head(4)
```

```python
# We need not more than 3 actors casted in the movie. We'll extract just 3 names from Cast feature by writing another function
```

```python
def fetchVal2(text):
    L=[]
    counter = 0
    for i in ast.literal_eval(text):
        if counter < 3:
            L.append(i['name'])
            counter = counter + 1
    return L
```

```python
movies['cast'] = movies['cast'].apply(fetchVal2)
```

```python
movies.head(2)
```

```python
# From 'crew' feature, we need only the name of the Director. We need to fetch the name where the key called 'job' has 
# value as 'Director'
```

```python
def fetchDirector(text):
    L=[]
    for i in ast.literal_eval(text):
        if i['job'] == 'Director':
            L.append(i['name'])
    return L
```

```python
movies['crew'] = movies['crew'].apply(fetchDirector)
```

```python
movies.head(4)
```

```python
#Now we need to remove spaces within names in genres, keywords,cast,crew features so that they can be tokenised properly
```

```python
def removeSpaces(text):
    L = []
    for i in text:
        L.append(i.replace(" ",""))
    return L
```

```python
movies['genres'] = movies['genres'].apply(removeSpaces)
movies['keywords'] = movies['keywords'].apply(removeSpaces)
movies['cast'] = movies['cast'].apply(removeSpaces)
movies['crew'] = movies['crew'].apply(removeSpaces)
```

# NLP concept -Bag of Words

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features = 5000, stop_words = 'english', binary = True)
```

```python
tagsVector = cv.fit_transform(movies['tags']).toarray()
```

```python
tagsVector.shape
```

```python
tagsVector
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
cosSimilar = cosine_similarity(tagsVector)
```

```python
cosSimilar.shape
```

```python
cosSimilar[0]
```

```python
list(enumerate(cosSimilar[0]))
```

```python
sorted(list(enumerate(cosSimilar[0])),reverse = True, key=lambda x : x[1])
```

```python
def recommend(cinema):
    
    #find index of the input cinema from movies dataset
    cinIndex = movies[movies['title']==cinema].index[0]
    
    #find the similar movies to that in the above index by using cosine similarity. lambda function is to sort the second column i.e.,cosSimilar values
    distances = sorted(list(enumerate(cosSimilar[cinIndex])),reverse = True,key = lambda x:x[1])

    #We need to recommend 6 similar movies, hence distances[1:6]. Ignoring 0th index as it is same as the input cinema
    for i in distances[1:6]:
        print(movies.iloc[i[0]].title)
```

```python
recommend("Avatar")
```

```python
recommend("Spider-Man")
```

```python
recommend("Pirates of the Caribbean: At World's End")
```
